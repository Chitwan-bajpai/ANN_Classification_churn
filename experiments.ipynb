```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
## Load the dataset
data=pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
#   PRPROCESS THE DATA
# drop irreleveant features
data=data.drop(['RowNumber','CustomerId','Surname'],axis=1)
data.head()
```

```python
# dealing with categorical data
label_encoder=LabelEncoder()
data['Gender']=label_encoder.fit_transform(data['Gender'])
```

```python
data
```

```python
# male and female converted into 0 and 1
# one hot encoder
from sklearn.preprocessing import OneHotEncoder
one_hot_geo=OneHotEncoder()
geo_one_hot_encoded=one_hot_geo.fit_transform(data[['Geography']])
# here i used two [] brackets in geogrsaphy beacause one hot encoder expecgts a 2d array
geo_one_hot_encoded
```

```python
geo_subColumns=one_hot_geo.get_feature_names_out(['Geography'])
```

```python
geo_data=geo_one_hot_encoded.toarray()
geo_data
```

```python
geo_encoded_df=pd.DataFrame(geo_data,columns=geo_subColumns)
geo_encoded_df
```

```python
## Combine all the one hot encoder columns with all the data
data=pd.concat([data.drop('Geography',axis=1),geo_encoded_df],axis=1)
data.head()
```

```python
# now we do not have any categorical features left all are converted into numerical features
```

Saving encoders and scalers in pickle file

```python
## SAVE THE ENCODERS AND SCALER
with open('label_encoder.pkl','wb') as file:
    pickle.dump(label_encoder,file)

with open('one_hot_geo.pkl','wb') as file:
    pickle.dump(one_hot_geo,file)
    
```

```python

# Divide the dataset into independent and dependent features
x=data.drop('Exited',axis=1)
y=data['Exited']

## Split the data in training and testing
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

## Scale these features
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

ANN IMPLEMENTATION

```python
import tensorflow  as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime

```

```python
## Build our ANN model
x_train.shape
```

Initialize a sequential ann model

```python
# we have 12 columns that means 12 initial inputs 
```

```python
model = Sequential([
    Dense(64, activation='relu', input_shape=(12,)),  # first hidden layer connected with input layer
    Dense(32, activation='relu'),  # hidden layer 2
    Dense(1, activation='sigmoid')  # output layer
])
```

```python
model.summary()
```

```python
# parameters are combination of all the weights and bias
```

Creating optimizers for backward propagation.

```python
import tensorflow
opt= tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss=tensorflow.keras.losses.BinaryCrossentropy()
loss
```

Above we created a binary cross-entropy loss function which measures how far model's prediction are from the actual values( basically y-y^) for binary classification problems.

# Compiling the model
When we call model.compile, we tell tensorflow the rules of how the model will learn
1. How to measure error(loss function)
2. How to improve optimizers- "How should i adjust my weights to get better"
3. What to track(evaluation metrices)

```python
# Compile the model
# model.compile(optimizer,loss,metrics)
model.compile(optimizer=opt,loss="binary_crossentropy",metrics=['accuracy'])
```

Setup the Tensorboard

```python

from tensorflow.keras.callbacks import EarlyStopping,TensorBoard

log_dir="logs/fit"+ datetime.datetime.now().strftime("%Y%m%d - %H%M%S")
tensorflow_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
# Set up Earlystopping
# why?
# When we are training a neural network we can train it for multiple epochs
# but there can be cases when our model is trained at its best at 20 epochs even it have 100 epochs
# it is not necessary to train it further
```

# Set up Earlystopping
## why?
 When we are training a neural network we can train it for multiple epochs
 but there can be cases when our model is trained at its best at 20 epochs even it have 100 epochs
 it is not necessary to train it further

```python
early_stopping_callback=EarlyStopping(monitor='val_loss',patience=20,restore_best_weights=True)
## here patience  is the no. of the epochs
```

NOW we will train the model

```python
history=model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs=100,
    callbacks=[tensorflow_callback,early_stopping_callback]
)
```

This code trains my neural network.
model.fit()- The main training function that:
. Feeds trainig data through the model
. Calculates loss
. Updates weights via backpropagation
. Repeats for the specified number of epochs

The model learns from 
validation_data=(x_test,y_test)
test data used to evaluate performance after each epoch.
